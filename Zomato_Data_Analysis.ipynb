```python
# this is second dataset for data analysis..
df_Country=pd.read_excel('Country-Code.xlsx')
df_Country
```

```python
df.columns
```

```python
# merging Zomato.csv and Country-Code.xlsx..

final_df=pd.merge(df,df_Country,on='Country Code',how='left')
```

```python
## to chek Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
# value counts on Country with index wise..

country_names=final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val=final_df.Country.value_counts().values
```

```python
country_val
```

```python
## pie Chart - top 4 Countries they use zomato
plt.pie(country_val[:4],labels=country_names[:4],autopct='%1.2f%%',colors=['pink','orange','blue','green'],explode = (0.05, 0.05, 0.09,0.017))
plt.title('Top 4 Countries.')
```

Observation:Zomato maximum records or transaction are from India

```python
# groupby function using for customer satisfaction and rating..
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
Ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating Count'})
```

```python
Ratings
```

```python
Ratings.head()
```

```python
# ploting Rating count on ratings..
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='rating Count',data=Ratings)
```

```python
# ploting color wise ratings 
sns.barplot(x='Aggregate rating',y='rating Count',hue='Rating color',data=Ratings,palette=['purple','red','orange','yellow','green','green'])
```

Observation:
1. Not Rated Count is very high -->0.0
2. Maximum No of Rating Range 2.5 to 3.4

```python
## Count Plot
# This plot represent frequency of ratings..
sns.countplot(x='Rating color',data=Ratings,palette=['purple','red','orange','yellow','green','green'])
```

```python
plt.figure(figsize = (8, 4))
sns.countplot(x = "Has Table booking", data = final_df, palette = "viridis",)
plt.title("RESTAURANT HAS TABLE BOOKING")
plt.show()
```

The data indicate that the majority of restaurants do not offer the option for advance table booking.

```python
# finging top 10 resturant based on votes count
votes=final_df.groupby("Restaurant Name")["Votes"].sum().reset_index()
votes=votes.sort_values(by="Votes",ascending=False)
votes.head(10)
```

```python
# finging top 10 resturant based on votes count visualization.
votes=final_df.groupby("Restaurant Name")["Votes"].sum().reset_index()
votes=votes.sort_values(by="Votes",ascending=False)
sns.barplot(x="Votes",y="Restaurant Name",data=votes[:10])
plt.title("TOP 10 RESTAURANTS BASED ON VOTES")
sns.set_style('darkgrid')
```
